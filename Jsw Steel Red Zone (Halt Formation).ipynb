```python
# Get Trips Function
def get_trips(query):
    df = spark.sql(query).toPandas()
    return df

# Basic cleaning and filter operations
def clean_trips(df):

    print('cleaning trips...', len(df))

    datetime_cols = ['created_at', 'trip_closure_time']
    for col in datetime_cols:
        df[col] = pd.to_datetime(df[col], errors='coerce')
        # df[col] = df[col].dt.tz_localize('Asia/Kolkata')

    df['src'] = df.apply(lambda row : [round(float(row['loading_lat']), 3), round(float(row['loading_long']),3)], axis = 1)
    df['dest'] = df.apply(lambda row : [round(float(row['unloading_lat']),3), round(float(row['unloading_long']),3)], axis = 1)
    df = df.drop(columns = ['loading_lat', 'loading_long', 'unloading_lat', 'unloading_long'])
    df['month'] = df['created_at'].dt.month
    
    df['trip_id'] = df['trip_id'].astype(int)
    df['device_id'] = df['device_id'].astype(str)
    df['year'] = df['created_at'].dt.year
    print("cleaned trips : ", len(df))
    return df.sort_values(by = 'created_at', ascending = True).reset_index(drop = True)
    
# Get pings from API
def get_pings(df):
    df = df.drop_duplicates(subset=['trip_id', 'device_id'])
    data_list = []
    required_columns = ['device_id', 'device_time', 'lat', 'lng']
    fetched_from_v1, fetched_from_v2 = 0, 0
    for index, row in tqdm(df.iterrows()):
        try:
            time.sleep(0.25)
            device_id = row['device_id']
            trip_created_at = row['created_at']
            if trip_created_at > datetime(2024, 10, 1):
                url = "http://yoda.freighttiger.com/v1/device/" + device_id + "/locations/trail?valid_locations=true&order=desc&invalid_locations=false"
                fetched_from_v1 += 1
            else:
                encoded_from_time = urllib.parse.quote(str(row['created_at']))
                encoded_to_time = urllib.parse.quote(str(row['trip_closure_time']))
                base_url = "http://yoda-v2.int.freighttiger.com/v1/device"
                url = f"{base_url}/{device_id}/locations/trail?from_time={encoded_from_time}&to_time={encoded_to_time}&order=DESC&valid_locations=true"
                # print(url)
                fetched_from_v2 += 1
            response = requests.get(url)
            data_list.extend(response.json()['data'])
        except Exception as e:
            print("Error : ", e, " for device : ", device_id)
    print("fetched from v1 : ", fetched_from_v1, " , fetched from v2 : ", fetched_from_v2)
    df = pd.DataFrame(data_list)
    if df.empty:
        return df
    else:
        return df[required_columns]

# Basic formatting for ping data
def clean_pings(df):
    # try:
    #     df['attributes'] = df['attributes'].apply(lambda x : ast.literal_eval(x))
    # except:
    #     # print("ping attribute formatting not required")
    #     pass
    # Convert from milliseconds to datetime in UTC timezone
    df['device_time_ist'] = pd.to_datetime(df['device_time'], unit='ms', utc=True)
    # Convert from UTC to IST (Asia/Kolkata)
    df['device_time_ist'] = df['device_time_ist'].dt.tz_convert('Asia/Kolkata')
    # Add Loc Column
    df['loc'] = df.apply(lambda row : [float(row['lat']), float(row['lng'])], axis=1)
    return df

# Fetch pings from analytics db
def fetch_location_data(df):
    trip_ids = list(set(df['trip_id']))
    # device_ids = list(set(df['device_id']))
    print("fetching locations from analytics db ...")
    project_data = ['device_id', 'device_time', 'lat', 'lng']
    project_data = {ele : 1 for ele in project_data}
    pings_from_db = pd.DataFrame(list(pings_collection.find(
        {'company_id': company_id, 'trip_id': {'$in': trip_ids}}, project_data
    )))
    
    if pings_from_db.empty == False:
        print("trips where location is available : ", len(set(pings_from_db['device_id'])))
    else:
        pings_from_db = pd.DataFrame(columns = ['device_id', 'device_time', 'lat', 'lng'])
    
    print('fetching locations from yoda api ...')

    # Use set subtraction
    # trip_ids = list(set(trip_ids) - set(pings_from_db['trip_id']))
    
    data_not_in_db = df[~df['device_id'].isin(list(set(pings_from_db['device_id'])))]


    if data_not_in_db.empty == False:
        pings_from_yoda = pd.DataFrame(get_pings(data_not_in_db))
        if pings_from_yoda.empty == False:
            pings_from_yoda['device_id'] = pings_from_yoda['device_id'].astype(str)
            print("trips where location is available : ", len(set(pings_from_yoda['device_id'])))
            # Correct concat usage
            pings = pd.concat([pings_from_db, pings_from_yoda], ignore_index=True)
    
        else:
            pings_from_yoda = pd.DataFrame(columns = ['device_id', 'device_time', 'lat', 'lng'])
            print("Locations not available in Yoda!")
            pings = pings_from_db
    else:
        print("No locations to fetch from api")
        pings = pings_from_db

    return clean_pings(pings)

    

# Plot the trips on Map
def plot_trips(df, obj, origin, destination, theft_location, halts):

    # Create a base map centered on India
    india_map = folium.Map(location=theft_location, zoom_start=8)

    if df.empty == False:
        # Sort data by device_time and reset index
        df = df.sort_values(by='device_time', ascending=True).reset_index()
        
        # Plot SIM pings as a polyline (blue color)
        for index, row in df.iterrows():
            folium.Circle(location=row['loc'], radius=50, color='blue', fill=True, fill_color='blue', fill_opacity=0.6, popup=f'Time: {row["device_time"]}').add_to(india_map)
    

    
    folium.Circle(location=origin, radius=3000, color='red', fill=True, fill_opacity=0.6).add_to(india_map)
    folium.Circle(location=destination, radius=3000, color='red', fill=True, fill_opacity=0.6).add_to(india_map)
    folium.Circle(location=theft_location, radius=1000, color='black', fill=True, fill_opacity=0.6).add_to(india_map)
    folium.Circle(location=theft_location, radius=250, color='yellow', fill=True, fill_opacity=0.6).add_to(india_map)

    for halt in halts:
        folium.Marker(location=halt['mean_location'], radius=100, color='green', fill=True, fill_opacity=0.6).add_to(india_map)
        for halt_location in halt['locations']:
            folium.Circle(location = halt_location, radius=50, color='green', fill=True, fill_opacity=0.6).add_to(india_map)

    # Display the map in the notebook
    display(india_map)

# Plot Plant boundary on Map
def plot_plant_boundary(obj):
    # Inverting the coordinates (switching longitude and latitude)
    inverted_coordinates = []
    for polygon in obj['geofence']['coordinates'][0]:
        inverted_coordinates.append([polygon[1], polygon[0]])
    
    # Creating a map centered at the first point in the inverted coordinates
    india_map = folium.Map(location=inverted_coordinates[0], zoom_start=10)
    folium.Polygon(locations=inverted_coordinates, color='red', fill=True, fill_opacity=0.2).add_to(india_map)
    
    display(india_map)

# Halt formation funciton
# Function to calculate halts
def form_halts(df):
    if df.empty:
        return None
    # Sort by device time
    df = df.sort_values(by='device_time_ist', ascending=True).reset_index(drop = True)
    
    # Add previous location and time
    df['prev_loc'] = df['loc'].shift(1)
    df['prev_device_time'] = df['device_time_ist'].shift(1)
    df['time_diff'] = df.apply(lambda row : (row['device_time_ist'] - row['prev_device_time']).total_seconds()/3600, axis = 1)
    
    # Apply speed calculation
    df.at[0, 'time_diff'] = 10
    df.at[0, 'prev_loc'] = df.loc[0]['loc']
    df['speed'] = df.apply(lambda row : haversine_distance(row['prev_loc'], row['loc'])/row['time_diff'], axis=1)
    df.at[0, 'speed'] = 0

    # return df
    halts = []
    halt = None

    for i, row in df.iterrows():
        if row['speed'] <= 3:
            if not halt:  
                # Start a new halt
                halt = {
                    "start_time": row['device_time_ist'],
                    "locations": [row['loc']],
                    "end_time": row['device_time_ist']  # Initialize end time
                }
            else:
                # Extend the current halt
                halt['locations'].append(row['loc'])
                halt['end_time'] = row['device_time_ist']

        elif halt:
            # Calculate duration of the halt
            duration = (halt['end_time']-halt['start_time']).total_seconds()/60

            if duration >= 10:
                latitudes = [loc[0] for loc in halt['locations']]
                longitudes = [loc[1] for loc in halt['locations']]
            
                mean_location = (round(np.mean(latitudes), 5), round(np.mean(longitudes), 5))

                # Append halt to the list if duration >= 10 mins
                halts.append({
                    "duration": round(duration),
                    "start_time": halt['start_time'],
                    "end_time": halt['end_time'],
                    "locations": halt['locations'],
                    "mean_location": mean_location
                })
            
            # Reset halt
            halt = None

    # Check if there's an ongoing halt at the end
    if halt:
        duration = (halt['end_time']-halt['start_time']).total_seconds()/60
        if duration >= 10:
            latitudes = [loc[0] for loc in halt['locations']]
            longitudes = [loc[1] for loc in halt['locations']]
            
            mean_location = (round(np.mean(latitudes), 5), round(np.mean(longitudes), 5))

            halts.append({
                "duration": round(duration, 2),
                "start_time": halt['start_time'],
                "end_time": halt['end_time'],
                "locations": halt['locations'],
                "mean_location": mean_location
            })

    return halts


```

```python
# Cluster halts
def cluster_halts(halt_list):
    df = pd.DataFrame(halt_list)
    # Define DBSCAN parameters
    eps_km = 0.1  # Max distance between points (in km)
    eps = eps_km / 6371.0  # Convert km to radians for DBSCAN
    min_samples = 5  # Minimum halts to form a cluster

    # Apply DBSCAN clustering
    db = DBSCAN(eps=eps, min_samples=min_samples, algorithm='ball_tree', metric='haversine').fit(np.array(df['mean_location'].tolist()))
    df['cluster_label'] = db.labels_ 
    return process_halt_clusters(df)

# process halt clusters
def process_halt_clusters(df):
    df = df[df['cluster_label'] != -1]
    unique_labels = list(set(df['cluster_label']))
    result = []
    for label in tqdm(unique_labels):
        df_cluster = df[df['cluster_label'] == label]
        latitudes = [loc[0] for loc in df_cluster['mean_location']]
        longitudes = [loc[1] for loc in df_cluster['mean_location']] 
        mean_location = (round(np.mean(latitudes), 5), round(np.mean(longitudes), 5))
        result.append({
            'unloading_address' : unloading_location,
            'frequent_halt_location' : mean_location,
            'count_of_halts_detected' : len(df_cluster),
            'frequent_halt_location_address' : None, #get_osm_address(mean_location),
            'frequent_halt_location_nearby_ameneties' : None, #fetch_nearby_places(mean_location, 200),
            'destination_coordinates' : destination_coords,
            'is_near_origin' : True if haversine_distance(origin_coords, mean_location) <= 5 else False,
            'is_near_destination' : True if haversine_distance(destination_coords, mean_location) <= 5 else False,
        })
    return result

# Fetch Nearby Places
def fetch_nearby_places(coordinates, radius):
    time.sleep(0.1)
    overpass_url = "https://overpass-api.de/api/interpreter"

    # Overpass QL query to fetch nearby places within a radius of 100m
    overpass_query = f"""
    [out:json];
    node
      (around:{radius},{coordinates[0]},{coordinates[1]})
      ["amenity"];
    out body;
    """

    try:
        response = requests.get(overpass_url, params={"data": overpass_query}, timeout=10)
        response.raise_for_status()
        data = response.json()

        places = []
        for element in data.get("elements", []):
            tags = element.get("tags", {})  # Ensure tags exist
            if tags and "amenity" in tags:  # Only process valid amenities
                place_lat = element["lat"]
                place_lon = element["lon"]
                distance = geodesic(coordinates, (place_lat, place_lon)).meters

                places.append({
                    "name": tags.get("name", "Unknown"),  # Get name, fallback to "Unknown"
                    "type": tags.get("amenity", "Unknown"),  # Get amenity type
                    "lat": place_lat,
                    "lon": place_lon,
                    "distance_from_halt_location_meters": round(distance, 2),
                })
        
        # Sort by distance
        return sorted(places, key=lambda x: x["distance_from_halt_location_meters"])

    except requests.exceptions.RequestException as e:
        print(f"Request error: {e}")
        return []
    except ValueError as e:
        print(f"JSON decode error: {e}")
        return []

# Get Address using osm
def get_osm_address(coordinates):
    geolocator = Nominatim(user_agent="geoapi")
    location = geolocator.reverse(coordinates, exactly_one=True)
    return location.address if location else "Address not found"

# process report format
def format_report(halt_locations):
    df = pd.DataFrame(halt_locations)
    df['company_id'] = 1182
    df['branch_id'] = 2
    df.insert(0, 'created_at', datetime.utcnow())
    df[['nearby_amenities', 'nearest_amenity', 'nearest_amenity_distance']] = None, None, None
    for index, halt in df.iterrows():
        if halt['frequent_halt_location_nearby_ameneties']:
            nearby_amenties = list(set(pd.DataFrame(halt['frequent_halt_location_nearby_ameneties'])['type']))
            df.at[index, 'nearby_amenities'] = nearby_amenties
            df.at[index, 'nearest_amenity'] = nearby_amenties[0]
            df.at[index, 'nearest_amenity_distance'] = list(set(pd.DataFrame(halt['frequent_halt_location_nearby_ameneties'])['distance_from_halt_location_meters']))[0]
    return df.to_dict(orient = 'records')
```

```python
report_columns = ['vehicle_number', 'created_at', 'trip_id', 'consignee_name', 'driver_name', 'driver_numbers', 'transporter_name', 'shipment_number', 'material_code']

report_obj = {
        'trips_location_within_50m' : pd.DataFrame(),
        'trips_location_within_1000m' : pd.DataFrame(),
        'trips_halt_within_50m' : pd.DataFrame(),
        'trips_halt_within_1000m' : pd.DataFrame(),
        'untracked_trips' : pd.DataFrame()
    }

result = pd.DataFrame()
step = 500
num_list = list(range(0, len(trips) + step, step))
for i, ele in enumerate(num_list):
    save_trips_to_db = []
    print("*******************************************************************")
    print("On Index : ", ele+step)

    if i <= len(num_list) - 1: 
        # Clean Trips Data
        trips_copy = trips[ele:ele+step].copy()
        trips_copy[['locations', 'halts']] = None, None

    # Fetch Locations
    locations = fetch_location_data(trips_copy)
    if locations.empty == True:
        continue

    # trips where location data is not available
    untracked_trips = trips_copy[~trips_copy['device_id'].isin(list(locations['device_id']))][report_columns]
    print("Untracked Trips : ", len(untracked_trips))
    report_obj['untracked_trips'] = pd.concat([report_obj['untracked_trips'], untracked_trips], ignore_index=True)
    
    # Form Halts and process the data
    for index, trip in trips_copy.iterrows():
        theft_zone_trip = False
        trip_locations = locations[locations['device_id'] == trip['device_id']]
        trip_locations = trip_locations.sort_values(by='device_time_ist')
        
        if not trip_locations.empty:
            # trips_copy.at[index, 'locations'] = trip_locations

            # Generate and sort halts
            trip_halts = form_halts(trip_locations)
            trip_halts.sort(key=lambda x: x['start_time'])

            if trip_halts:
                # trips_copy.at[index, 'halts'] = trip_halts
                
                # Function to calculate halt count and duration
                def calculate_halts_within_range(halts, distance_threshold):
                    series_started = False
                    series_halts = []
                    first_halt_time = None
                    last_halt_time = None

                    for halt in halts:
                        distance = haversine_distance(halt['mean_location'], theft_location)

                        if distance <= distance_threshold:
                            # Start series
                            if not series_started:
                                series_started = True
                                first_halt_time = halt['start_time']
                            
                            series_halts.append(halt)
                            last_halt_time = halt['end_time']

                        # End series if a halt is out of range
                        elif series_started:
                            break

                    if series_halts:
                        total_halts = len(series_halts)
                        total_duration = (last_halt_time - first_halt_time).total_seconds() / 60 if last_halt_time else 0
                        return total_halts, round(total_duration, 2)
                    return 0, 0

                # --- Check for halts within 50m ---
                if not theft_zone_trip:
                    total_halts, total_duration = calculate_halts_within_range(trip_halts, 0.05)
                    if total_halts > 0:
                        theft_data = {"trip_id": trip['trip_id'], "total_halts": total_halts, "total_duration_minutes": total_duration, 'trip_halts' : trip_halts, 'trip_label' : 'trips_halt_within_50m'}
                        save_trips_to_db.append(theft_data)
                        theft_data = pd.DataFrame([theft_data]).merge(trips_copy[trips_copy['trip_id'] == trip['trip_id']][report_columns], on = 'trip_id', how = 'left')
                        report_obj['trips_halt_within_50m'] = pd.concat([report_obj['trips_halt_within_50m'], theft_data], ignore_index= True)
                        theft_zone_trip = True
                
                # --- Check for halts within 1000m ---
                if not theft_zone_trip:
                    total_halts, total_duration = calculate_halts_within_range(trip_halts, 1)
                    if total_halts > 0:
                        theft_data = {"trip_id": trip['trip_id'], "total_halts": total_halts, "total_duration_minutes": total_duration, 'trip_halts' : trip_halts, 'trip_label' : 'trips_halt_within_1000m'}
                        save_trips_to_db.append(theft_data)
                        theft_data = pd.DataFrame([theft_data]).merge(trips_copy[trips_copy['trip_id'] == trip['trip_id']][report_columns], on = 'trip_id', how = 'left')
                        report_obj['trips_halt_within_1000m'] = pd.concat([report_obj['trips_halt_within_1000m'], theft_data], ignore_index= True)
                        theft_zone_trip = True

            # --- Check for individual locations within 50m ---
            if not theft_zone_trip:
                series_started = False
                series_locations = []
                first_time = None
                last_time = None

                for _, location in trip_locations.iterrows():
                    distance = haversine_distance(location['loc'], theft_location)

                    if distance <= 0.05:
                        if not series_started:
                            series_started = True
                            first_time = location['device_time_ist']
                        
                        series_locations.append(location)
                        last_time = location['device_time_ist']
                    elif series_started:
                        break

                if series_locations:
                    total_locations = len(series_locations)
                    total_duration = (last_time - first_time).total_seconds() / 60 if last_time else 0
                    theft_data = {"trip_id": trip['trip_id'], "total_halts": total_locations, "total_duration_minutes": total_duration, 'trip_label' : 'trips_location_within_50m'}
                    save_trips_to_db.append(theft_data)
                    theft_data = pd.DataFrame([theft_data]).merge(trips_copy[trips_copy['trip_id'] == trip['trip_id']][report_columns], on = 'trip_id', how = 'left')
                    report_obj['trips_location_within_50m'] = pd.concat([report_obj['trips_location_within_50m'], theft_data], ignore_index= True)
                    theft_zone_trip = True

            # --- Check for individual locations within 1000m ---
            if not theft_zone_trip:
                series_started = False
                series_locations = []
                first_time = None
                last_time = None

                for _, location in trip_locations.iterrows():
                    distance = haversine_distance(location['loc'], theft_location)

                    if distance <= 1:
                        if not series_started:
                            series_started = True
                            first_time = location['device_time_ist']
                        
                        series_locations.append(location)
                        last_time = location['device_time_ist']

                    elif series_started:
                        break

                if series_locations:
                    total_locations = len(series_locations)
                    total_duration = (last_time - first_time).total_seconds() / 60 if last_time else 0
                    theft_data = {"trip_id": trip['trip_id'], "total_halts": total_locations, "total_duration_minutes": total_duration, 'trip_label' : 'trips_location_within_1000m'}
                    save_trips_to_db.append(theft_data)
                    theft_data = pd.DataFrame([theft_data]).merge(trips_copy[trips_copy['trip_id'] == trip['trip_id']][report_columns], on = 'trip_id', how = 'left')
                    report_obj['trips_location_within_1000m'] = pd.concat([report_obj['trips_location_within_1000m'], theft_data], ignore_index= True)
                    theft_zone_trip = True
    # Saving all trips separately --> Not Required
    # result = pd.concat([result, trips_copy], ignore_index=True)
    reports_collection.insert_many(save_trips_to_db)
    print("Data Saved to DB.")
```
